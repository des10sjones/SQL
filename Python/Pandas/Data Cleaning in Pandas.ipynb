# Cleaning Data using Pandas

```python
import pandas as pd
```

```python
df = pd.read_excel(r"C:\Users\DestinJones\Desktop\SQL\Python\Data\Customer Call List.xlsx")
df
```

```python
df = df.drop_duplicates()
df
```

```python
#Drops Column
df = df.drop(columns ='Not_Useful_Column')
    
```

```python
df
```

```python
df['Last_Name'] = df['Last_Name'].str.strip('123._/')
```

```python
df
```

```python
# df['Last_Name'] = df['Last_Name'].str.lstrip('/')
# df['Last_Name'] = df['Last_Name'].str.rstrip('_')
# df['Last_Name'] = df['Last_Name'].str.strip('123._/')
```

```python
df
```

```python
df['Phone_Number'] = df['Phone_Number'].str.replace('[^a-zA-Z0-9]','',regex = True)
df['Phone_Number']
```

```python
df['Phone_Number'] = df['Phone_Number'].apply(lambda x: str(x))
```

```python
df['Phone_Number'] = df['Phone_Number'].apply(lambda x: x[0:3] + '-' + x[3:6] + '-' + x[6:10])
```

```python
df
```

```python
df['Phone_Number'] = df['Phone_Number'].str.replace('nan--','')
df['Phone_Number'] = df['Phone_Number'].str.replace('Na--','')
df
```

```python
df[['Street_Address','State','Zip_Code']] = df["Address"].str.split(',',expand = True)
df
```

```python
df['Paying Customer'] = df['Paying Customer'].str.replace('Yes','Y')

```

```python
df['Paying Customer'] = df['Paying Customer'].str.replace('No','N')
```

```python
df
```

```python
df['Paying Customer'] = df['Paying Customer'].str.replace('Y','Yes')
df['Paying Customer'] = df['Paying Customer'].str.replace('N','No')
df['Paying Customer'] = df['Paying Customer'].str.replace('No/a','N/a')
```

```python
df
```

```python
df['Do_Not_Contact'] = df['Do_Not_Contact'].str.replace('Yes','Y')
df['Do_Not_Contact'] = df['Do_Not_Contact'].str.replace('No','N')
df['Paying Customer'] = df['Paying Customer'].str.replace('Yes','Y')
df['Paying Customer'] = df['Paying Customer'].str.replace('No','N')
df
```

```python
df = df.replace('N/a', '')
```

```python
df
```

```python
df = df.fillna('')
df
```

```python
for x in df.index:
    if df.loc[x, 'Do_Not_Contact'] == 'Y':
        df.drop(x, inplace=True)
df
```

```python
df['Do_Not_Contact'] = df['Do_Not_Contact'].replace('','N')
df
```

```python
for x in df.index:
    if df.loc[x, 'Phone_Number'] == '':
        df.drop(x, inplace=True)
df

# We couldve done df.dropna(subset="Phone_Number",inplace=True)
```

```python
df.reset_index(drop = True)
```

```python
df.to_csv("Customer Call List Cleaned", index=False)
```

```python
df.to_json("Customer Call List Cleaned.json", orient="records", indent=4)
```
