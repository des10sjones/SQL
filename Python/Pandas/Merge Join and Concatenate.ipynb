# Merge, Join, and Concatenate

```python
import pandas as pd
```

```python
df1 = pd.read_csv(r"C:\Users\DestinJones\Desktop\SQL\Python\Data\LOTR.csv")
df1
```

```python
df2 = pd.read_csv(r"C:\Users\DestinJones\Desktop\SQL\Python\Data\LOTR 2.csv")
df2
```

```python
df1.merge(df2)
```

```python
df1.merge(df2, how = 'inner')
```

```python
df1.merge(df2, how = 'inner',on = 'FellowshipID')
```

```python
df1.merge(df2, how = 'inner',on = ['FellowshipID','FirstName'])
```

```python
df1.merge(df2, how = 'outer')
```

```python
df1.merge(df2, how = 'left')
```

```python
df1.merge(df2, how = 'right')
```
